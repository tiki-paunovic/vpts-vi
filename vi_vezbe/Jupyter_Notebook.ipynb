### Literalno programiranje([Donald Ervin Knuth](https://www-cs-faculty.stanford.edu/~knuth/faq.html), 1984)

Literalno programiranje je paradigma programiranja koju je uveo Donald Knut 1984. godine. Njegova ideja je bila da se raunarski program posmatra kao literalno delo u cilju njegove bolje dokumentovanosti. Smatrao je da ne treba da mislimo da nam je glavni cilj da dajemo instrukcije raunaru ta da radi, nego da radije treba da se skoncentriemo na to da objasnimo ljudima ta elimo raunar da uradi. Tako dolazimo do toga da se programski kod nalazi unutar opisnog teksta, to je potpuno suprotno od onoga sa ime ste se do sada susretali (tekst unutar koda).

### Mathematica ([Stephen Wolfram](https://www.stephenwolfram.com/), 1988)

Idejni tvorac programa *Mathematica* za simbolika matematika izraunavanja je Stiven Volfram, britanski fiziar, matematiar, programer, autor i biznismen, poznat po brojnim radovima iz podruja fizike elementarnih estica, matematike, kosmologije, teorije sloenosti, kao i po sajtu [Volfram alfa](https://www.wolframalpha.com/). *Wolfram Mathematica* se sastoji iz dva dela, iz jezgra i prednjeg kraja. Jezgro interpretira kod pisan u Volfram jeziku i vraa rezultat a prednji kraj poseduje grafiko radno okruenje koje je u obliku interaktivne radne sveske. Dizajnirao ga je [Theodore Gray](http://home.theodoregray.com/) i ono je ujedno prvi istorijski primer radnih sveski. Radne sveske se sastoje od niza elija koje sadre programski kod sa automatskom bojom za isticanje sintakse, formatirani tekst zajedno sa izlaznim rezultatima, ukljuujui matematike izraze, grafiku, grafike komponente, tabele i zvuk. Svi sadraji i formatiranje su generisani automatski ili menjani interaktivno. Takoe su podrane standardne mogunosti obrade rei. Radne sveske sa njihovim sadrajima se mogu prezentovati kao izrazi koji mogu biti kreirani, menjani ili analizirani u programu Matematika ili konvertovani u druge formate. Dokumenti se takoe mogli prikazati kao slideshow za prezentacije.

### Maple (1992)

Prva verzija programe *Maple* za simbolina i brojana izraunavanja sa grafikim korisnikim okruenjem je izala 1989. godine (prva verzija Maple-a izala 7 godina ranije) dok je prva verzija sa interaktivnom radnom sveskom izala 1992. godine. I ovde se takoe kombinuju kod, tekst i izlazni rezultati izvrenog koda. Podrana je i grafika. Dok se u programu Matematika (kao i u ostalim programima sa radnim okruenjem oblika radne sveske koje emo nabrojati, pa i u samoj Jupyter radnoj svesci) kod izvrava kombinacijom tastera \< Shift \> i \< Enter \>, u programu Maple se izvrava pritiskom na taster \< Enter \>.

### Sage ([William Stein](https://wstein.org/), 2005)

Pojavom softvera otvorenog koda (1998), dovode do razvoja mnogih besplatnih programa otvorenog koda kao i do rasta popularnosti Python naunih alata otvorenog koda.  
  
Python je u devedesetim godinama prolog veka proiren tako da mu je dodan tip nizova za numerika izraunavanja pod nazivom *Numeric*. U 2001. godini Travis Olifant, Peru Piterson i Erik Dons su zajednikim naporima napravili standardnu Python biblioteku numerikih operacija na vrhu *Numeric* strukture niza podataka i nazvali dobijenu biblioteku *SciPy*. *SciPy* koriste naunici, analitiari i inenjeri za razna nauna i tehnika izraunavanja. Inae, Travis Olifant je ovek koji je 2006. godine napisao biblioteku *NumPy* kao meavinu modula *Numeric* i *Numarray*. *NumPy* je Python biblioteka sa velikim kolekcijom matematikih funkcija koja podrava rad sa velikim, viedimenzionalnim nizovima i matricama.

Ubrzo nakon toga, Fernando Perez objavio je *IPython* (interaktivni Python), okruenje komandne linije za interaktivna izraunavanja u Python-u, sa podrkom za vizualizaciju podataka te paralelno i distribuirano izraunavanje. Na taj nain je sebi napravio *mini-Mathematica* Python program nad tipovima *Numeric* i *Gnuplot* sa nekih 250+ linija koda.

*John Hunter* je 2003. godine objavio prvu verziju *Matplotlib*-a, 2D Python biblioteke za crtanje grafikona. Ono to je zanimljivo je to da je *Matplotlib* nastao kao zakrpa za *IPython*, koju Fernando Perez nije imao vremena da pregleda i odobri zbog rada na doktoratu.

*Sage* je besplatan program otvorenog koda za simbolina raunarska izraunavanja koji je nastao 2005. godine kao alternativa za programe  *Magma*, *Maple*, *Mathematica* i *MATLAB*. Velika prednost mu je ta to je besplatan, za razliku od ova etiri nabrojana programa. Baziran je na gore nabrojanim Python tipovima i bibliotekama kao i jo mnogim drugim, a napisan je u programskim jezicima *Python* i *Cython*. Sage je interaktivna radna sveska a korisnikovo radno okruenje je internet pregleda.


### IPython ([Fernando Prez](https://bids.berkeley.edu/people/fernando-p%C3%A9rez), 2011)

*IPython* je postao interaktivna radna sveska 2011. godine. Korisnikovo radno okruenje je takoe internet pregleda. *Fernando Prez* i [*Robert Kern*](https://www.enthought.com/person/robert-kern/) su radili na interaktivnoj radnoj radnoj svesci istovremeno dok je razvijana Sejd radna sveska. *Fernando Prez* i *William Stein* bili su u kontaktu od 2005. godine pa nadalje i zajedno su saraivali s obzirom da je *IPython* korien u *Sage* okruenju komandne linije.

### Jupyter Notebook (2015)


Projekat *Jupyter* je nastao 2014. godine. Projekat odrava rad u nauci o podacima (engl. *data science*) kao i nauna izraunavanja u vie programskih jezika te razvoj programa otvorenog koda. Baziran je na *IPython*-u koji ini zadnji kraj tj. jezgro Jupyter interaktivne radne sveske. Kao zadnji kraj Jupyter radne sveske podrana su jezgra za pedesetak programskih jezika. Naziv *Jupyter* potie od tri osnovna programska jezika koja su bila podrana u projektu: ***Ju**lia*, ***Pyt**hon* i ***R***. Naziv je takoe dat u ast Galilejovih radnih svesaka u kojima je opisao otkria Jupiterovih meseca. O prednjem kraju tj. interaktivnoj *IPython* radnoj svesci smo ve priali. *IPython* radno okruenje je preimenovano u *Jupyter notebook* 2015. godine (IPython 4.0  Jupyter 1.0). Vrlo brzo je Jupyter radna sveska po popularnosti prestigla Matematika radnu svesku.

### Formatiranje teksta 

*Markdown* je jednostavan jezik za oznaavanje koji nam pomae da obian tekst formatiramo tako da izgleda mnogo bolje. *Markdown* su kreirali [*John Gruber*](https://en.wikipedia.org/wiki/John_Gruber) i [*Aaron Swartz*](https://en.wikipedia.org/wiki/Aaron_Swartz) 2004. godine.

Kada piemo tekst u eliju koji formatiramo sa *Markdown* sintaksom, mi dodajemo atribute na rei, a izgled teksta vidimo tek nakon izvravanja elije. Da bismo pripremili eliju za pisanje teksta sa *Markdown* sintaksom, treba da pritisnemo kombinaciju tastera  \< Esc \> i \< M \>. Ako elimo da elija ponovo bude spremna za pisanje i izvravanje koda, treba da pritisnemo kombinaciju tastera  \< Esc \> i \< Y \>.
